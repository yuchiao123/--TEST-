```python
# 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# 
iris = load_iris()

#  DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

#  species 
species_map = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
df['species'] = df['species'].map(species_map)

# 
plt.figure(figsize=(10, 6))
for species in df['species'].unique():
    species_df = df[df['species'] == species]
    plt.scatter(species_df['sepal length (cm)'], species_df['sepal width (cm)'], label=species)

plt.xlabel(' (cm)')
plt.ylabel(' (cm)')
plt.title('')
plt.legend()
plt.show()

# 
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.boxplot([df.loc[df['species'] == 'Setosa','sepal length (cm)'],
             df.loc[df['species'] == 'Versicolor','sepal length (cm)'],
             df.loc[df['species'] == 'Virginica','sepal length (cm)']],
             labels=['Setosa', 'Versicolor', 'Virginica'])
plt.title('')
plt.ylabel(' (cm)')

plt.subplot(1, 2, 2)
plt.boxplot([df.loc[df['species'] == 'Setosa','sepal width (cm)'],
             df.loc[df['species'] == 'Versicolor','sepal width (cm)'],
             df.loc[df['species'] == 'Virginica','sepal width (cm)']],
             labels=['Setosa', 'Versicolor', 'Virginica'])
plt.title('')
plt.ylabel(' (cm)')

plt.tight_layout()
plt.show()

# 
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(df['sepal length (cm)'], bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('')
plt.xlabel(' (cm)')
plt.ylabel('')

plt.subplot(1, 2, 2)
plt.hist(df['sepal width (cm)'], bins=10, alpha=0.7, color='blue', edgecolor='black')
plt.title('')
plt.xlabel(' (cm)')
plt.ylabel('')

plt.tight_layout()
plt.show()
```
